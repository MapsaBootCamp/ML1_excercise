```python
titanic_df.drop('Cabin', axis=1, inplace=True)
```

```python
titanic_df.drop(['Name', 'Ticket', 'PassengerId'], axis=1 , inplace=True)
```

```python
titanic_df.groupby('Pclass').mean().Age
```

```python
#df[column list].apply(your_fun , axis=1)


def clean_age(colum):
    age = colum[0]
    p_class = colum[1]
    
    if pd.isnull(age):
        if p_class == 1 :
            return 38
        elif p_class == 2:
            return 29
        else:
            return 25
    else:
        return age
    
titanic_df['Age'] = titanic_df[['Age', 'Pclass']].apply(clean_age, axis=1)
```

```python
titanic_df
```

```python
titanic_df.Sex = pd.Categorical(titanic_df.Sex)
```

```python
titanic_df.Sex = titanic_df.Sex.cat.codes
titanic_df
titanic_df.Embarked = pd.Categorical(titanic_df.Embarked)
titanic_df.Embarked = titanic_df.Embarked.cat.codes
```

```python
titanic_df
```

```python
titanic_df.info()
```

```python
titanic_df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = titanic_df.iloc[: , 1:]
y = titanic_df['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import Perceptron
```

```python
lm_classifier = Perceptron(penalty='l2')
```

```python
lm_classifier.fit(X_train, y_train)
```

```python
y_predict = lm_classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_predict))
```
