```python
import pandas as pd
```

```python
heart_disease_dataset=pd.read_csv('Heart_Disease_Dataset.csv')
heart_disease_dataset
```

```python
heart_disease_dataset.dtypes
```

```python
heart_disease_dict={}
for col in heart_disease_dataset:
    heart_disease_dict.update({col:heart_disease_dataset[col].unique().tolist()})
print(heart_disease_dict)
```

```python
heart_disease_dataset=heart_disease_dataset.drop(heart_disease_dataset.loc[heart_disease_dataset['number_of_major_vessels(ca)'] == '?'].index)
heart_disease_dataset=heart_disease_dataset.drop(heart_disease_dataset.loc[heart_disease_dataset['thal'] == '?'].index)
heart_disease_dataset.shape
```

```python
heart_disease_dataset['number_of_major_vessels(ca)']=heart_disease_dataset['number_of_major_vessels(ca)'].astype('float64')
heart_disease_dataset['thal']=heart_disease_dataset['thal'].astype('float64')
```

```python
heart_disease_dataset.dtypes
```

```python
heart_disease_dataset[heart_disease_dataset['diagnosis_of_heart_disease(num)']!=0]=1
heart_disease_dataset[heart_disease_dataset['diagnosis_of_heart_disease(num)']==0]=0
heart_disease_dataset
```

```python
from sklearn.utils import shuffle
heart_disease_dataset = shuffle(heart_disease_dataset, random_state=20)
heart_disease_dataset
```

```python
X=heart_disease_dataset.iloc[:,:-1]
y=heart_disease_dataset['diagnosis_of_heart_disease(num)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=20)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussian_model = GaussianNB()
gaussian_model.fit(X_train_scaled, y_train)
y_pred=gaussian_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
# Confusion matrix
print("Confusion Matrix")
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
# Classification Report
print("\nClassification Report")
report = classification_report(y_test, y_pred)
print(report)
# Accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print('GaussianNB Classification Accuracy of the model: {:.2f}%'.format(accuracy*100))
```
