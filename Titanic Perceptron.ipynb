```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.tail()
```

```python
titanic = titanic[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked', 'class', 'who', 'adult_male', 
                   'deck', 'embark_town', 'alive', 'alone']].apply(lambda x: x.astype('category').cat.codes)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
```

```python
x = titanic.iloc[:, :-1]
y = titanic['survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=88)
```

```python
perceptron = Perceptron(penalty='l2')
```

```python
pipe_steps = [('scaler', preprocessing.StandardScaler()), ('per', perceptron)]
pipeline =Pipeline(steps=pipe_steps)
pipeline.fit(x_train, y_train)
y_predict = pipeline.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_predict))
```
