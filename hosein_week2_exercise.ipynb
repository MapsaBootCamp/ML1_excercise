<div id="bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
		<p></p>
        برای این مجموعه داده ها قصد داریم با دو فرمول زیر مقادیر آنها را تغییر دهیم:
<br>
۱. اگر عددی کمتر از هشت و نیم است با فرمول زیر آن عدد تغییر کند:
<br>
$$ y = 10 + \frac{x}{10} $$
<br>
۲. اگر عددیبیشتر یا مساوی هشت و نیم است با فرمول زیر آن عدد تغییر کند :
<br>
$$ y =  -0.0051x^{2} + 0.8586x + 3.4332 $$
<br>
توجه کنید که حتی بعد از تغییر باید بیشینه عدد نهایتا بیشینه عدد موجود در داده فعلی باشد. 25 درصد داده های بزرگتر را در آرایه زیر بریزید
</div>

```python
def first_func(x):
    return 10. + x/10.
```

```python
def second_func(x):
    return 3.4332 + x * (0.8586 - 0.0051 * x)
```

```python
num = np.piecewise(numbers, [numbers<8],[first_func,second_func])
```

```python
limit = numbers.max()
new_numbers = np.where(num < limit,num,numbers)
```

```python
new_75 = np.percentile(new_numbers,74)
```

```python
top_new_twenty_five = new_numbers[new_numbers > new_75].round(2)
```

```python
top_new_twenty_five.sort()
```

```python
top_new_twenty_five.size
```

<div id="bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
		<p></p>
 25 درصد داده های کوچکتر آرایه اول را در آرایه زیر بریزید
</div>

```python
old_25 = np.percentile(numbers,26)
```

```python
down_old_twenty_five = numbers[numbers < old_25]
```

```python
down_old_twenty_five.sort()
```

<div id="bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
      <p></p>
        داده های متناظر دو آرایه به دست آمده را از یکدیگر کسر کنید و در یک آرایه سه بعدی که اندازه ابعاد سوم و دوم آن یک میباشد بریزید
</div>

```python
down_old_twenty_five.size
```

```python
diff = np.zeros(min(down_old_twenty_five.size, top_new_twenty_five.size))
```

```python
for i,(top,down) in enumerate(zip(top_new_twenty_five, down_old_twenty_five)):
    diff[i] = top - down
```

```python
diff
```
