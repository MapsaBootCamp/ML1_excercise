```python
import pandas as pd
```

```python
football_dataset=pd.read_csv('/content/UEFA_CL_Player_stats.csv')
football_dataset.head()
```

```python
# finding top goalscorer
goal_df = (football_dataset.groupby(['name','season year'])['goals'].sum()).reset_index(name ='total_goals')
goal_df=goal_df.sort_values(by=['season year','total_goals'],ascending=False)
season_years=set(goal_df['season year'])
top_goalscorer=pd.DataFrame(columns=['name','season year','total_goals'])
for season_year in season_years:
  df=goal_df.loc[goal_df['season year']==season_year]
  df=df.reset_index(drop=True)
  result=df.loc[df['total_goals']==(df['total_goals'].max())]
  frames=[top_goalscorer,result]
  top_goalscorer = pd.concat(frames,ignore_index=True)
print(top_goalscorer.sort_values(by=['season year'],ignore_index=True))
```

```python
teams=set(football_dataset['team'])
football_dataset_changed=pd.DataFrame(columns=['team','season','season year','total_goals','total_yellow_cards','total_red_cards'])
for season_year in season_years:
  df_season=football_dataset.loc[football_dataset['season year']==season_year]
  df_season=df_season.reset_index(drop=True)
  # print(df_season['season'])
  for team in teams:
    df=df_season.loc[df_season['team']==team]
    df=df.reset_index(drop=True)
    football_dataset_changed=football_dataset_changed.append({'team': team,'season':df_season.loc[0,'season'],'season year':season_year, 'total_goals': df['goals'].sum(), 'total_yellow_cards': df['yellowCards'].sum(),'total_red_cards':df['redCards'].sum()}, ignore_index=True)

football_dataset_changed.to_csv('/content/football_dataset_changed.csv')
football_dataset_changed
```
