```python
import numpy as np
import pandas as pd

cancer_df = pd.read_csv('breast-cancer.data',names=["Class", "age", "menopause", "tumor-size","inv-nodes","node-caps", 
                                                    "deg-malig","breast","breast-quad", "irradiat"], header=None)
cancer_df
```

```python
cancer_df.info()
```

```python
cancer_df['deg-malig']= cancer_df['deg-malig'].map(str)
cancer_df['age']= cancer_df['age'].map(str)
cancer_df.info()
```

```python
cancer_df.apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)
```

Features are almost independant.

```python
data = cancer_df.drop(columns='Class')
target = cancer_df['Class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(OrdinalEncoder(handle_unknown='ignore'), CategoricalNB())
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, prediction))
confusion_matrix(y_test, prediction)
```
