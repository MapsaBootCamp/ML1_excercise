		<p></p>
        برای این مجموعه داده ها قصد داریم با دو فرمول زیر مقادیر آنها را تغییر دهیم:
<br>
۱. اگر عددی کمتر از هشت و نیم است با فرمول زیر آن عدد تغییر کند:
<br>
$$ y = 10 + \frac{x}{10} $$
<br>
۲. اگر عددیبیشتر یا مساوی هشت و نیم است با فرمول زیر آن عدد تغییر کند :
<br>
$$ y =  -0.0051x^{2} + 0.8586x + 3.4332 $$
<br>
توجه کنید که حتی بعد از تغییر باید بیشینه عدد نهایتا بیشینه عدد موجود در داده فعلی باشد. 25 درصد داده های بزرگتر را در آرایه زیر بریزید
</div>

```python
new_numbers =list(map(lambda x: 10+(x/10) if x<8.5 else ((-0.0051*x**2)+(0.8586*x)+3.4332),numbers))
new_numbers = np.array(new_numbers)
quarter4_new = np.quantile(new_numbers,0.75)

top_new_twenty_five = np.array(list(filter(lambda x: x>=quarter4_new,new_numbers)))
top_new_twenty_five
```

<div id="bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
		<p></p>
 25 درصد داده های کوچکتر آرایه اول را در آرایه زیر بریزید
</div>

```python
median_old = np.median(numbers)
quarter2_old = np.quantile(numbers,0.25)

down_old_twenty_five = np.array(list(filter(lambda x: x<=quarter2_old, numbers)))
down_old_twenty_five
```

</div>

```python
sub_np = top_new_twenty_five - down_old_twenty_five
sub_np.reshape(1,1,24)
```
