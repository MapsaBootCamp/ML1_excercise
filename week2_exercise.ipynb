		<p></p>
        برای این مجموعه داده ها قصد داریم با دو فرمول زیر مقادیر آنها را تغییر دهیم:
<br>
۱. اگر عددی کمتر از هشت و نیم است با فرمول زیر آن عدد تغییر کند:
<br>
$$ y = 10 + \frac{x}{10} $$
<br>
۲. اگر عددیبیشتر یا مساوی هشت و نیم است با فرمول زیر آن عدد تغییر کند :
<br>
$$ y =  -0.0051x^{2} + 0.8586x + 3.4332 $$
<br>
توجه کنید که حتی بعد از تغییر باید بیشینه عدد نهایتا بیشینه عدد موجود در داده فعلی باشد. 25 درصد داده های بزرگتر را در آرایه زیر بریزید
</div>

```python
def less(numbers):
    max_numbers=numbers.max()
    for i in range(len(numbers)):
        if numbers[i] < 8.5:
            new_num=10+(numbers[i]/10)
            if new_num > max_numbers:
                numbers[i]=max_numbers
            else:
                numbers[i]=new_num
        else:
            new_num=((-0.0051)*(numbers[i]**2))+(0.8586*numbers[i]+3.4332)
            if new_num > max_numbers:
                numbers[i]=max_numbers
            else:
                numbers[i]=new_num 
    return numbers

def Sortby_Third_Quantile(numbers):
    numbers=less(numbers)
    numbers.sort()
    quantile_number=np.quantile(numbers,0.75)
    top_new_twenty_five=numbers[numbers>quantile_number]
    return top_new_twenty_five
```
