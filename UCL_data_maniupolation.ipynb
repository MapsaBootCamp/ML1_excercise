```python
import numpy as np
import pandas as pd
```

```python
UL_df = pd.read_csv('UEFA_CL_Player_stats.csv')
print(list(UL_df.columns))
UL_df
```

```python
UCL_df = UL_df[['name', 'season year', 'team', 'goals', 'yellowCards', 'redCards']]
UCL_df
```

```python
Top_scorer = pd.DataFrame(UCL_df.groupby('season year').apply(lambda x: x.sort_values(by=['goals'], ascending=False).head(1)))
Top_scorer[['season year', 'name', 'goals', 'team']]
```

```python
teams_goals = pd.DataFrame(UCL_df.groupby(['season year', 'team']).goals.sum()).sort_values(by=['season year', 'goals'], ascending=False)
teams_goals
```

```python
teams_YCards = pd.DataFrame(UCL_df.groupby(['season year', 'team']).yellowCards.sum()).sort_values(by=['season year', 'yellowCards'], ascending=False)
teams_YCards
```

```python
teams_RCards = pd.DataFrame(UCL_df.groupby(['season year', 'team']).redCards.sum()).sort_values(by=['season year', 'redCards'], ascending=False)
teams_RCards
```

```python
Output_df = pd.DataFrame(UCL_df.groupby(['team', 'season year']).goals.sum()).sort_values(by=['team', 'goals'], ascending=False)
redCard_col = pd.DataFrame(UCL_df.groupby(['team', 'season year']).redCards.sum()).sort_values(by=['team', 'redCards'], ascending=False)
yellowCard_col =  pd.DataFrame(UCL_df.groupby(['team', 'season year']).yellowCards.sum()).sort_values(by=['team', 'yellowCards'], ascending=False)
# redCard_col
Output_df
# yellowCard_col
```

```python
Output_df['yellowCards'] = yellowCard_col['yellowCards']
Output_df['redCards'] = redCard_col['redCards']
Output_df
```

```python
Output_df.to_csv('UCL_Teams.csv')
```
