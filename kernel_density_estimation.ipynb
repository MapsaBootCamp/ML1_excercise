```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
```

```python
# generate data
samples = np.random.normal(0., 10., 10000)
h,e = np.histogram(samples, bins=100, density=True)
x = np.linspace(e.min(), e.max())

# plot the histogram
plt.figure(figsize=(10,7),dpi=100)
plt.bar(e[:-1], h, width=np.diff(e), edgecolor='black', align='edge',label='histogram')

# plot the kde
kde = sts.gaussian_kde(samples)
plt.plot(x, kde.pdf(x), color='red',linewidth=2.5 ,label='kde')

plt.title('Kernel Density Estimation')
plt.legend()
plt.show()
```
