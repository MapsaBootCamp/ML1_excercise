```python
import pandas as pd
# get Breast_Cancer_Dataset from UCI
breast_cancer_dataset=pd.read_csv('Breast_Cancer_Dataset.csv')
breast_cancer_dataset.head()
```

```python
breast_cancer_dataset.shape
```

```python
breast_cancer_dataset.dtypes
```

```python
breast_cancer_dict={}
for col in breast_cancer_dataset:
    breast_cancer_dict.update({col:breast_cancer_dataset[col].unique().tolist()})

# breast_cancer_dataset.astype('object').describe(include='all').loc['unique', :]
print(breast_cancer_dict)
```

```python
breast_cancer_dataset=breast_cancer_dataset.drop(breast_cancer_dataset.loc[breast_cancer_dataset['node-caps'] == '?'].index)
breast_cancer_dataset=breast_cancer_dataset.drop(breast_cancer_dataset.loc[breast_cancer_dataset['breast-quad'] == '?'].index)
breast_cancer_dataset.shape
```

```python
breast_cancer_dataset_modified=breast_cancer_dataset.reset_index(drop=True)
breast_cancer_dataset_modified
```

```python
features = breast_cancer_dataset_modified.columns.tolist()
features.remove('Class')
features.remove('irradiat')
print(features) 
```

```python
print('=' * 30)
for var in features:
    print(breast_cancer_dataset_modified[var].value_counts().sort_values(ascending=False) / breast_cancer_dataset_modified.shape[0])
    print('=' * 30)
    print()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
data_encoded = encoder.fit_transform(breast_cancer_dataset_modified[features])
breast_cancer_dataset_encoded = pd.DataFrame(data_encoded, columns=features)
breast_cancer_dataset_encoded
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
target_encoded = encoder.fit_transform(breast_cancer_dataset_modified['irradiat'])
breast_cancer_dataset_encoded['irradiat'] = target_encoded
breast_cancer_dataset_encoded
```

```python
from sklearn.utils import shuffle
breast_cancer_dataset_encoded = shuffle(breast_cancer_dataset_encoded, random_state=20)
breast_cancer_dataset_encoded
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_dataset_encoded.drop('irradiat', axis=1), breast_cancer_dataset_encoded['irradiat'], test_size=0.3, random_state=20)
```

```python
from sklearn.naive_bayes import CategoricalNB
gategorical_nb_model = CategoricalNB()
gategorical_nb_model.fit(X_train, y_train)
y_pred=gategorical_nb_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
# Confusion matrix
print("Confusion Matrix")
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
# Classification Report
print("\nClassification Report")
report = classification_report(y_test, y_pred)
print(report)
# Accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print('CategoricalNB Classification Accuracy of the model: {:.2f}%'.format(accuracy*100))
```

```python
target_count = breast_cancer_dataset_encoded['irradiat'].value_counts()
print('Class 0:', target_count[0])
print('Class 1:', target_count[1])
```

```python
# Class count
count_class_0, count_class_1 = breast_cancer_dataset_encoded.irradiat.value_counts()

# Divide by class
df_class_0 = breast_cancer_dataset_encoded[breast_cancer_dataset_encoded['irradiat'] == 0]
df_class_1 = breast_cancer_dataset_encoded[breast_cancer_dataset_encoded['irradiat'] == 1]
```

```python
#Random over-sampling
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
breast_cancer_dataset_over = pd.concat([df_class_0, df_class_1_over], axis=0)
```

```python
target_count = breast_cancer_dataset_over['irradiat'].value_counts()
print('Random over-sampling:')
print('Class 0:', target_count[0])
print('Class 1:', target_count[1])
```

```python
breast_cancer_dataset_over = shuffle(breast_cancer_dataset_over, random_state=20)
```

```python

X_train_over, X_test_over, y_train_over, y_test_over = train_test_split(breast_cancer_dataset_over.drop('irradiat', axis=1), breast_cancer_dataset_over['irradiat'], test_size=0.3, random_state=20)
```

```python
gategorical_nb_over = CategoricalNB()
gategorical_nb_over.fit(X_train_over, y_train_over)
y_pred_over=gategorical_nb_over.predict(X_test_over)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print("+++++++  over sampling  +++++++")
# Confusion matrix
print("Confusion Matrix")
matrix = confusion_matrix(y_test_over, y_pred_over)
print(matrix)
# Classification Report
print("\nClassification Report")
report = classification_report(y_test_over, y_pred_over)
print(report)
# Accuracy of the model
accuracy = accuracy_score(y_test_over, y_pred_over)
print('CategoricalNB Classification Accuracy of the model: {:.2f}%'.format(accuracy*100))
```

```python
#Random under-sampling
df_class_0_under = df_class_0.sample(count_class_1)
breast_cancer_dataset_under = pd.concat([df_class_0_under, df_class_1], axis=0)
```

```python
target_count = breast_cancer_dataset_under['irradiat'].value_counts()
print('Random Under-sampling:')
print('Class 0:', target_count[0])
print('Class 1:', target_count[1])
```

```python
breast_cancer_dataset_under = shuffle(breast_cancer_dataset_under, random_state=20)
```

```python
X_train_under, X_test_under, y_train_under, y_test_under = train_test_split(breast_cancer_dataset_under.drop('irradiat', axis=1), breast_cancer_dataset_under['irradiat'], test_size=0.3, random_state=20)
```

```python
gategorical_nb_under = CategoricalNB()
gategorical_nb_under.fit(X_train_under, y_train_under)
y_pred_under=gategorical_nb_under.predict(X_test_under)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print("+++++++  under sampling  +++++++")
# Confusion matrix
print("Confusion Matrix")
matrix = confusion_matrix(y_test_under, y_pred_under)
print(matrix)
# Classification Report
print("\nClassification Report")
report = classification_report(y_test_under, y_pred_under)
print(report)
# Accuracy of the model
accuracy = accuracy_score(y_test_under, y_pred_under)
print('CategoricalNB Classification Accuracy of the model: {:.2f}%'.format(accuracy*100))
```
