<div id="bayes" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
		<p></p>
        برای این مجموعه داده ها قصد داریم با دو فرمول زیر مقادیر آنها را تغییر دهیم:
<br>
۱. اگر عددی کمتر از هشت و نیم است با فرمول زیر آن عدد تغییر کند:
<br>
$$ y = 10 + \frac{x}{10} $$
<br>
۲. اگر عددیبیشتر یا مساوی هشت و نیم است با فرمول زیر آن عدد تغییر کند :
<br>
$$ y =  -0.0051x^{2} + 0.8586x + 3.4332 $$
<br>
توجه کنید که حتی بعد از تغییر باید بیشینه عدد نهایتا بیشینه عدد موجود در داده فعلی باشد. 25 درصد داده های بزرگتر را در آرایه زیر بریزید
</div>

```python
len_numbers=numbers.size
temp=np.zeros(len_numbers)
for count in range(len_numbers):
    if numbers[count]<8.5:
        num=10+(numbers[count]/10)
        if num < numbers[count]:
            temp[count]=num
        else:
            temp[count]=numbers[count]
    else:
        num=(-0.0051*(numbers[count]**2))+(0.8586*numbers[count])+3.4332
        if num < numbers[count]:
            temp[count]=num
        else:
            temp[count]=numbers[count]
numbers_v2=temp.copy()
numbers_v2
```

```python
#top_new_twenty_five = 
temp=np.sort(temp)
print(temp)
quarter=int(len_numbers*0.25)
top_new_twenty_five = temp[:(3*quarter):-1]
top_new_twenty_five=np.sort(top_new_twenty_five)
top_new_twenty_five
```

 25 درصد داده های کوچکتر آرایه اول را در آرایه زیر بریزید

```python
#down_old_twenty_five = 
down_old_twenty_five = temp[:quarter]
down_old_twenty_five
```

</div>
