```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic.head()
```

```python
titanic.shape
Age = titanic['Age'].tolist()
Ticket =  titanic['Ticket'].tolist()
Fare =  titanic['Fare'].tolist()
Pclass =  titanic['Pclass'].tolist()
Parch =  titanic['Parch'].tolist()

plt.figure(figsize=(16,12))
plt.subplot(221)
plt.hist(Age, bins=40)
plt.title('Age Histogram')
plt.subplot(222)
plt.hist(Parch, bins=12)
plt.title('Parch Histogram')
plt.subplot(223)
plt.hist(Fare, bins=70)
plt.title('Fare Histogram')
plt.subplot(224)
plt.hist(Pclass, bins=5)
plt.title('Pclass Histogram')
```

```python
pie_counts = titanic['Pclass'].value_counts()
survived = titanic['Survived'].value_counts()
plt.figure(figsize=(12,7))
plt.subplot(121)
plt.pie(pie_counts, labels=['class_3', 'class_2', 'class_1'], autopct='%1.1f%%', startangle=90, shadow=True)
plt.subplot(122)
plt.pie(survived, labels=['murdred', 'survived'], autopct='%1.1f%%', startangle=90, shadow=True)
```

```python
plt.figure(figsize=(17,7))
plt.subplot(121)
sns.boxplot(x="Survived", y="Fare", hue="Sex", data=titanic)
plt.subplot(122)
sns.boxplot(x="Pclass", y="Fare", hue="Survived", data=titanic)
```
